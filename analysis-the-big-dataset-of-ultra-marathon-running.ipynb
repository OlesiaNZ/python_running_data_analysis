```python
# Drop rows with missing values
df.dropna(subset=['Event distance/length'], inplace=True)

# Filtering rows where Event distance/length contains '50'
df[df['Event distance/length'].str.contains('50')]['Event distance/length'].unique() # See lot's of dirty data
```

```python
# Count the occurrences of each unique value
unique_values = df[df['Event distance/length'].str.contains('50')]['Event distance/length'].unique()

# Create a dictionary to store the counts
counts = {}

# Iterate over the unique values and count occurrences
for value in unique_values:
    count = df[df['Event distance/length'] == value].shape[0]
    counts[value] = count

# Print the counts
for value, count in counts.items():
    print(f"{value}: {count} occurrences")
```

```python
# I see the dirty data in Event distance/length column as 50miles, 150km/3Etappen, 50h, 50.9 mi, 50K, 50Km, 250km/6stages, 150km/4 Etappen, 16:50h, 50Mile, 50Miles, 50mi+
```

```python
# Remove space between values
df['Event distance/length'] = df['Event distance/length'].str.replace(' ', '')
# Check if it works. Last time I saw '50.9 mi'
df[df['Event distance/length'].str.contains('50.9')]['Event distance/length'].unique() # Cool, no space between
```

```python
# Transform xMile, xMiles to xmi
df['Event distance/length'] = df['Event distance/length'].str.replace(r'\b(\d+)\s*Miles?\b', r'\1mi', regex=True)
```
